```python
# Count a charcter in a given string


# using count function directly
s = input("enter string = ")
serach = input ("Enter what you want to search for = ")

print(s.count(serach))

# Using loop for
counter = 0
for i in s:
  if i == serach:
    counter += 1
print("Length of char is =",counter)
```

```python
# count the length of string without using the length function

s = input("enter string = ")

count = 0
for i in s:
  count += 1
print("Length of string is =",count)
```

```python
# Write a program to remove a particular char from a string

s = input("enter string = ")
serach = input ("Enter what you want to remove = ")

result = ""

for i in s:
  if i != serach:
    result += i
print(result)
```

```python
# Write a program to find the number of words in a string without using split functionn

s = input("enter string = ")

L = []
term = ''

for i in s:

  if i != ' ':
    term = term + i
  else:
    L.append(term)
    term = ''

L.append(term)
print(L)
```

```python
# Write a program to convert the first word of each letter into uppercase without using title function

s = input("enter string = ")

L = []

for i in s.split():
  L.append(i[0].upper() + i[1:].lower())

print(" ".join(L))
```

```python
# Write a program to convert intergers into string without using str function

number = int(input("enter the number = "))

digits = "0123456789"
result = ''
while number != 0:
  result = digits[number % 10] + result
  number = number//10
print(result)
```
